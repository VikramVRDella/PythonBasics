```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6])
y=np.array([0,250])

plt.plot(x,y)
plt.show()
```

```python
print("Plotting with Line....")
a=np.array([1,5])
b=np.array([1,30])
plt.plot(a,b)
plt.show()
```

```python
print("Plotting without Line......")
c=np.array([1,8])
d=np.array([3,23])
plt.plot(c,d, 'o')
plt.show()
```

```python
print("Multiple Points...")
f=np.array([1,3,5,2])
g=np.array([2,4,6,8])
plt.plot(f,g)
plt.show()
```

```python
print("Default X Points....")
h=np.array([3,2,6,5,8,1])
plt.plot(h)
plt.show()
```

```python
plt.plot(h,marker='*')
plt.show
```

```python
plt.plot(h,linestyle='dotted')
plt.show()
```

```python
plt.plot(h,linestyle='dashed')
plt.show()
```

```python
plt.plot(h,ls=':',color='r')
plt.show()
```
