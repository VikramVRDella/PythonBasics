```python
"""
Importing Required Libraries
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
"""
Creating Sample Values...
"""
temperature=[20,30,15,25,40]
sales=[15,20,15,25,50]

x=np.array([temperature]).T
y=np.array(sales)
```

```python
"""
Importing LinearRegression Model
"""
lmodel=LinearRegression()
lmodel=lmodel.fit(x,y)
lmodel_slope=lmodel.coef_
lmodel_intercept=lmodel.intercept_
print(lmodel_slope)
print(lmodel_intercept)
```

```python
y_predict=lmodel.predict(x)
rmse=np.sqrt(mean_squared_error(y,y_predict))
r2=lmodel.score(x,y)
print("Mean Square Error :",rmse)
print("R Square :",r2)
```

```python
plt.scatter(temperature,sales,edgecolors='r',marker='*')
plt.plot(temperature,y_predict)
```
